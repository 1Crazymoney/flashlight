```python
%pylab inline
import numpy
```

```python
def plot(path):
    words = []
    w_len = []
    c_len = []
    with open(path, "r") as f:
        for line in f:
            l = line.strip()
            w_len.append(len(l.split(" ")))
            c_len.append(len(l))
            words += l.split(" ")
    plt.hist(w_len)
    plt.show()
    plt.hist(c_len)
    print("words", numpy.mean(w_len), numpy.std(w_len), numpy.max(w_len))
    print("chars", numpy.mean(c_len), numpy.std(c_len))
    print(len(set(words)))
```

```python
plot("/datasets01/ptb/060817/words/ptb.train.txt")
```

```python
plot("/datasets01/ptb/060817/words/ptb.valid.txt")
```

```python
plot("/datasets01/ptb/060817/words/ptb.test.txt")
```
