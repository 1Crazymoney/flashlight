```python
path = "/checkpoint/antares/experiments/fl_new_master/augposemb/vit/"

def plot_one(path, name, valid=True):
    val_epoch = []
    val_top1 = []
    with open(path + "/antares.err", "r") as f:
        for line in f:
            if (valid and "Validation Loss" in line):
                val_epoch.append(int(line.strip().split("Epoch ")[1].split(" ")[0]))
                val_top1.append(float(line.strip().split(": ")[-1]))
            elif (not valid and "Train Loss" in line):
                val_epoch.append(len(val_top1) * 100)
                val_top1.append(float(line.strip().split(": ")[-1]))
    plt.plot(val_epoch, val_top1, label=name)
```

```python
plt.figure(figsize=(20, 8))
plt.grid()
plt.xlabel("epoch")
plt.ylabel("MNIST loss")

plot_one(path + "mnist_baseline_use_pos_embfalse_use_aug_pos_embtrue_use_aug_pos_emb_augtrue", "CAPE", False)
plot_one(path + "mnist_baseline_use_pos_embfalse_use_aug_pos_embtrue_use_aug_pos_emb_augfalse", "sin2d", False)
plot_one(path + "mnist_baseline_use_pos_embfalse", "nopos", False)
plot_one(path + "mnist_baseline", "abspos", False)

plt.legend()
# plt.ylim(2.5, 4.5)
```
